# K-Means Clustering

## Importing the libraries

```
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

## Importing the dataset

```
dataset = pd.read_csv('Mall_Customers.csv')
X = dataset.iloc[:, [3,4]].values
```

## Using the elbow method to find the optimal number of clusters

```
from sklearn.cluster import KMeans
wcss = []

for i in range(1,11):
  K_mean = KMeans(n_clusters = i, init = 'k-means++',random_state=42)
  K_mean.fit(X)
  wcss.append(K_mean.inertia_)
  # print(wcss)

plt.plot(range(1,11),wcss)
plt.title("Elbow-Method")
plt.xlabel('No of clusters')
plt.ylabel('WCSS')
plt.show()
```

## Training the K-Means model on the dataset

```
K_mean = KMeans(n_clusters = 5, init = 'k-means++',random_state=42)
y_kmeans = K_mean.fit_predict(X)
```

```
print(y_kmeans)
```

## Visualising the clusters

```
plt.scatter(X[y_kmeans == 0,0],X[y_kmeans == 0,1],s=100,c='red',label = 'Cluster 1')
plt.scatter(X[y_kmeans == 1,0],X[y_kmeans == 1,1],s=100,c='blue',label = 'Cluster 2')
plt.scatter(X[y_kmeans == 2,0],X[y_kmeans == 2,1],s=100,c='green',label = 'Cluster 3')
plt.scatter(X[y_kmeans == 3,0],X[y_kmeans == 3,1],s=100,c='cyan',label = 'Cluster 4')
plt.scatter(X[y_kmeans == 4,0],X[y_kmeans == 4,1],s=100,c='violet',label = 'Cluster 5')
plt.scatter(K_mean.cluster_centers_[:,0],K_mean.cluster_centers_[:,1],s=300,label = 'Cluster Centres')
plt.legend()
plt.title('K_Means')
plt.xlabel('Income')
plt.ylabel('Spending')
plt.show()
```
